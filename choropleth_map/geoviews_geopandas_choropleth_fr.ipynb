## Premi√®res cartes avec geoviews

on importe les outils dont nous aurons besoin, et on initialise geoviews : 

```python
import geoviews as gv
import geoviews.feature as gf
import xarray as xr
from cartopy import crs

gv.extension('bokeh')
```

`geoviews.feature` contient la description g√©om√©trique des diff√©rentes r√©gions, comme les terres, l'oc√©an, et les fronti√®res des pays. Affichons tout √ßa : 

```python
(gf.land * gf.ocean * gf.borders).opts(
    'Feature', 
    projection=crs.Mercator(), 
    global_extent=True, 
    width=500
)
```

Avec cartopy, nous pouvons facilement changer la [projection](https://scitools.org.uk/cartopy/docs/latest/crs/projections.html): 

```python
(gf.land * gf.ocean * gf.borders).opts(
    'Feature', 
    projection=crs.NearsidePerspective(central_longitude=0, 
                                       central_latitude=45), 
    global_extent=True, 
    width=500, height=500
)
```

## Charger un fichier GeoJSON avec pandas

Les fichiers [GeoJSON](https://geojson.org/) encodent les donn√©es g√©ographiques des r√©gions, et peuvent √™tre facilement manipul√©s avec [geopandas](http://geopandas.org/index.html). 

En cherchant "france geojson"sur Google, je suis tomb√© sur [france-geojson](https://github.com/gregoiredavid/france-geojson), un repo maintenu par Gr√©goire David, que je remercie chaleureusement. Depuis ce repo, j'ai t√©l√©charg√© [departements-version-simplifiee.geojson](https://github.com/gregoiredavid/france-geojson/blob/master/departements-version-simplifiee.geojson) en tant que `france-geojson/departements-version-simplifiee.geojson`. 

Faites de m√™me, puis importez geopandas pour lire ce fichier: 

```python
import geopandas as gpd
sf = gpd.read_file('france-geojson/departements-version-simplifiee.geojson')
sf.head()
```

Nous voyons que la g√©om√©trie de chaque d√©partement est stock√©e comme un polygone, avec la longitude et la latitude de chacun des points du polygone. 

Il est facile de dessiner un d√©partement avec geopandas : il nous suffit de s√©lectionner une ligne dans la dataframe (ici la premi√®re pour le d√©partement de l'Ain), et d'√©valuer la g√©om√©trie : 

```python
sf.loc[0].geometry
```

Nous allons maintenant tracer l'ensemble des d√©partements, avec une couleur indiquant une valeur mesur√©e dans chaque d√©partement. Pour l'instant, nous n'avons pas encore de valeur √† utiliser, donc nous en engendrons une al√©atoirement : 

```python
import numpy as np
sf['value'] = np.random.randint(1, 10, sf.shape[0])
```

Regardons ce bout de code de plus pr√®s. 

√Ä droite, on utilise numpy pour engendrer un tableau d'entiers entre 1 et 10, de longeur √©gale au nombre de lignes dans la dataframe, c'est √† dire `sf.shape[0]`. Ensuite, on assigne ce tableau √† une nouvelle colonne dans la dataframe, appel√©e `value`.

## Affichage de donn√©es g√©ographiques avec geoviews


Dans le style de holoviews, il nous faut juste deux lignes pour afficher les d√©partements. La premi√®re d√©clare le jeu de polygones √† partir de notre dataframe geopandas. Nous choisissons explicitement le nom du d√©partement et la valeur comme "dimensions de valeur" (les dimensions correspondant √† des valeurs mesur√©es, d√©pendant du d√©partement): 

```python
deps = gv.Polygons(sf, vdims=['nom','value'])
```

Ensuite, nous affichons les d√©partements avec quelques options. La plus importante est `color=dim('value')`. Elle veut simplement dire que nous allons mapper `value` vers une √©chelle de couleurs lin√©aire pour chaque d√©partement. Et vous pouvez trouver une explication compl√®te de ce que sont les objets `dim` dans la [documentation de holoviews](http://holoviews.org/user_guide/Style_Mapping.html).

**Veuillez noter que la colonne utilis√©e pour la couleur doit √™tre dans `vdims`. J'oublie souvent!**

```python
from geoviews import dim
deps.opts(width=600, height=600, toolbar='above', color=dim('value'),
          colorbar=True, tools=['hover'], aspect='equal')
```

Nous avons utilis√© bokeh comme backend, et nous b√©n√©ficions donc des outils usuels de cette librairie (au dessus de la figure, comme je l'ai demand√©). Vous pouvez essayer de zoomer avec la molette de votre souris, de d√©placer la carte, et de survoler les d√©partements pour obtenir des informations d√©taill√©es. 

## Cr√©ez une carte choropl√®the avec vos propres donn√©es

Dans l'exemple simple ci-dessus, nous avons d√©fini al√©atoirement une valeur pour chaque d√©partement. Mais en pratique, vous allez vouloir afficher des donn√©es qui ne sont pas pr√©sentes dans la dataframe geopandas. Nous avons donc besoin d'associer des donn√©es externes √† cette dataframe avant affichage. 

C'est ce que nous allons faire maintenant, en prenant comme exemple la population estim√©e dans chaque d√©partement en 2019. J'ai trouv√© ces donn√©es [ici](https://www.insee.fr/fr/statistiques/1893198) au format xls, et les ai converties en fichier tsv (tab-separated values) pour cr√©er le fichier [population_france_departements_2019.tsv](https://drive.google.com/open?id=1TxWTzMGgheODN2hFJD4NPeGSw1PnvEwA). 

Pour lire ce fichier, nous utilisons pandas (dans sa version classique). Nous n'avons en effet pas besoin de geopandas ici car ce dataset ne contient aucune information g√©ographique. Nous utilisons la m√©thode bien connue `read_csv`, en sp√©cifiant que la tabulation est utilis√©e pour s√©parer les valeurs d'une ligne, et que la virgule indique les milliers. Vous pouvez simplement ouvrir ce fichier texte pour vous en assurer. 

```python
import pandas as pd
df = pd.read_csv('population_france_departements_2019.tsv', sep='\t', thousands=',')
df.head()
```

Apr√®s les premi√®res colonnes qui indiquent le code et le nom de d√©partement, nous avons trois jeus de colonnes pour le nombre d'habitants du d√©partement. Les pr√©fixes correspondent √†: 

* `t_` : nombre total
* `h_` : nombre d'hommes 
* `f_` : nombre de femmes 

Et dans chaque jeu, il y a une colonne par tranche d'√¢ge. 

Maintenant, nous devons associer les donn√©es de cette dataframe pandas (`df`) √† notre dataframe geopandas (`sf`). C'est tr√®s facile avec la m√©thode `merge`. 

Ci-dessous, on merge les donn√©es sur chaque ligne, en s'assurant que la valeur de la colonne `code`de `sf` est √©gale √† la valeur de la colonne `depcode` de `df`. L'option `suffixes` concerne les colonnes ayant le m√™me nom dans `df` et `sf`, comme `nom`. Cet option d√©termine le nom que les colonnes auront apr√®s le merge. Ici, je souhaite que la colonne qui vient de `sf` garde son nom, et que celle venant de `df` r√©cup√®re un suffixe `_y`.

```python
jf = sf.merge(df, left_on='code', right_on='depcode', suffixes=('','_y'))
jf.head(2)
```

Et nous pouvons maintenant afficher la population totale de chaque d√©partement. Nous ajoutons au passage le nombre d'hommes et le nombre de femmes aux tooltips, en ajoutant ces colonnes √† `vdims`. De plus, veuillez noter que j'ai normalis√© le nombre d'habitants pour l'afficher en millions. 

```python
# declare how the plot should be done
regions = gv.Polygons(jf, vdims=['nom', 't_total', 'h_total', 'f_total'])
regions.opts(width=600, height=600, toolbar='above', color=dim('t_total')/1e6, 
             colorbar=True, tools=['hover'], aspect='equal')
```
